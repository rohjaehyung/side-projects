# Merging Business file and Sentiment Analysis
From previous working csv file, I plan to merge the business profile json file. 

```python
import pandas as pd
import datetime

df = pd.read_csv('2022review.csv')

df.head()
```

Opening with Pandas library, as usual. And checking the data sets.

```python
business = pd.read_json('yelp_academic_dataset_business.json', lines=True)
business.info()
```

I've merged business profile dataset with review dataset by matching business id. <br>
The below chart shows cities with number of reviews in 2022.

```python
merged_df2 = pd.merge(business, df, on = 'business_id', how = 'left')
merged_df2 = merged_df2.dropna()
merged_df2.groupby('city').count()
```

Exploring data first by city. Then by Postal code, and by star_x a.k.a average business rating greater than 3.

```python
merged_df2[merged_df2['city']=='philadelphia']
```

```python
pc_19103 = merged_df2[merged_df2['postal_code']== '19103']
pc_19103[pc_19103['stars_x'] > 3]
```

The review of "The Love" seem to have good rating. Extracting only the reviews, as shown below.

```python
the_love = pc_19103[pc_19103['name'] == 'The Love']
the_love['text']
```

Using nltk library, the SentimentIntensityAnalyer allowed to evaluate the sentiment of the customers.

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
sid = SentimentIntensityAnalyzer()

scores  = []
for sentence in the_love['text']:
    score = {'sentence': sentence}
    score.update(sid.polarity_scores(sentence))
    scores.append(score)
    
df_scores = pd.DataFrame(scores)
df_scores
```

As shown above, the emphasis of LOVE and expressions increases the positive scores.
