# Review Data Clean and Drop

I import the review dataset again using Pandas library.

```python
import pandas as pd

df = pd.read_json('yelp_academic_dataset_review.json', lines =True)
df.head()
```

Checking data structure

```python
df.info()
```

Sorting by date descending, since I do not need old data.

```python
df.sort_values('date', ascending =False, inplace = True)
df
```

```python
import datetime
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

Dropping all the data except from 2022 by striping year, month, day by datetime library.

```python
df = df[df['year'] == 2022]
```

```python
df.to_csv('2022review.csv', index = False)
```
